```python
region_data = data.groupby('WHO Region').agg({
    'New cases': 'sum',
    'New deaths': 'sum',
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
    'Active': 'sum',
    
    'New cases': 'sum',
    'New deaths': 'sum',
    'New recovered': 'sum',
    '1 week % increase': 'mean'
}).reset_index()



print(region_data)


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.scatter(region_data['New cases'], region_data['New deaths'], c=region_data['Death Rate (%)'], cmap='viridis', s=100, edgecolor='k')
plt.colorbar(label='Death Rate (%)')
plt.xlabel('New Cases')
plt.ylabel('New Deaths')
plt.title('New Cases vs New Deaths by WHO Region')

for i, row in region_data.iterrows():
    plt.annotate(row['WHO Region'], (row['New cases'], row['New deaths']))

plt.show()
```

# Predictive Modeling

```python
from sklearn.model_selection import train_test_split


features = data[['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 'New deaths']]
target = data['1 week % increase']

#split data 
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error


model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)

# predictions
predictions = model.predict(x_test)
```

<h>model evaluation </h>

```python
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)

r2 = model.score(x_test, y_test)

print(f'MAE: {mae}')

print(f'MSE: {mse}')

print(f'R Score: {r2}')
```

<h>model tuning </h>

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(x_train, y_train)

best_model = grid_search.best_estimator_
```

# Plot the feature importances


```python
import matplotlib.pyplot as plt

# Feature importance
importances = best_model.feature_importances_
indices = np.argsort(importances)[::-1]
feature_names = features.columns

plt.figure(figsize=(10, 6))
plt.title("Feature Importance")
plt.bar(range(x_train.shape[1]), importances[indices], align="center")

plt.xticks(range(x_train.shape[1]), feature_names[indices], rotation=90)
plt.show()
```

# Calculate death rate and recovery rate
# Evaluate the current situation

```python

data['death_rate'] = (data['Deaths'] / data['Confirmed']) * 100
data['recovery_rate'] = (data['Recovered'] / data['Confirmed']) * 100


death_rate = data.loc[0, 'death_rate']
recovery_rate = data.loc[0, 'recovery_rate']

#
def Determine_state(death_rate, recovery_rate):
    
    if death_rate < 2 and recovery_rate > 70:
        return "Continue current strategy effictive "

    elif death_rate > 2 and recovery_rate < 70:
        return "higher risk rates"     #it mean this rejon very danger and need new strategy 
    
    else:
        return "Monitor closely"       #These regions are in a situation where current strategies are relatively effective


strategy_recommendation = Determine_state(death_rate, recovery_rate)
print(f"Strategy Recommendation: {strategy_recommendation}")
```

<h>it still chech all updated data to detect rejon risks across the day</h>

```python
import time

def monitor_and_adjust(data):
    while True:
        # Recalculate rates
        data['death_rate'] = (data['Deaths'] / data['Confirmed']) * 100
        data['recovery_rate'] = (data['Recovered'] / data['Confirmed']) * 100
        
        # Evaluate intervention for each region
        for index, row in data.iterrows():
            recommendation = Determine_state(row['death_rate'], row['recovery_rate'])
            region_name = row['Country/Region']
            print(f"Region {region_name}: {recommendation}")
        
        time.sleep(86400)  # 86400 seconds in a day
        

monitor_and_adjust(data)
```
